# **Tiền xử lý dữ liệu**
1. Hiển thị toàn bộ dữ liệu
2. Điền các giá trị bị thiếu của dữ liệu theo phương pháp trung bình với các trường có giá trị số và theo phương pháp tần số cao nhất với giá trị rời rạc
3. Chuẩn hóa dữ liệu theo min max
4. Liên tục hóa thuộc tính variety
5. Rời rạc hóa thuộc tính sepal.length



```python
import pandas as pd
import numpy as np
```

```python
#Hiển thị toàn bộ dữ liệu
data=pd.read_csv('FLOWER.csv')
data.head()
```

```python
#Điền các giá trị bị thiếu của dữ liệu theo phương pháp trung bình với các trường có giá trị số và theo phương pháp tần số cao nhất với giá trị rời rạc
data.value_counts()
data.isna().sum()
```

```python
# prompt: #với sepal.length	sepal.width	petal.length	petal.width ta sẽ điền vào mising value giá trị mean
df1=data.copy()
df1['sepal.length'].fillna(round(df1['sepal.length'].mean()), inplace=True)
df1['sepal.width'].fillna(round(df1['sepal.width'].mean()), inplace=True)
df1['petal.length'].fillna(round(df1['petal.length'].mean()), inplace=True)
df1['petal.width'].fillna(round(df1['petal.width'].mean()), inplace=True)
```

```python
#với variety ta sẽ thay thế các mising value bằng giá trị có tần số cao nhất
df1['variety'].fillna(data['variety'].mode()[0], inplace = True)
df1.isna().sum()
```

```python
#Chuẩn hóa dữ liệu theo min max
df2 = df1.copy()

sepal_length_max = df2['sepal.length'].max()
sepal_length_min = df2['sepal.length'].min()

sepal_width_max = df2['sepal.width'].max()
sepal_width_min = df2['sepal.width'].min()

petal_length_max = df2['petal.length'].max()
petal_length_min = df2['petal.length'].min()

petal_width_max = df2['petal.width'].max()
petal_width_min = df2['petal.width'].min()

# Chuẩn hóa

df2['sepal.length'] = (df2['sepal.length'] - sepal_length_min) / (sepal_length_max - sepal_length_min)
df2['sepal.width'] = (df2['sepal.width'] - sepal_width_min) / (sepal_width_max - sepal_width_min)
df2['petal.length'] = (df2['petal.length'] - petal_length_min) / (petal_length_max - petal_length_min)
df2['petal.width'] = (df2['petal.width'] - petal_width_min) / (petal_width_max - petal_width_min)

df2

```

```python
#Liên tục hóa thuộc tính variety
df3=df2.copy()
dummies=pd.get_dummies(df3['variety'])
df3.head()
df3=pd.concat([df3,dummies],axis=1)
df3.drop('variety',axis=1,inplace=True)
```

```python
#rời rạc hóa thuộc tính sepal.gth
df4 = df3.copy()

df4['sepal.length'] = pd.cut(df4['sepal.length'],
             bins = [-0.1, 0.2, 0.4, 0.6, 0.8, 1],
             labels = ["Very Low", "Low", "Medium", "High", "Very High"])
```

# **Xây dựng mô hình hồi quy tuyến tính để đoán kết quả của thuộc tính variety dựa trên các thuộc tính khác**

```python

df=df1.copy()

#chia ra thành train với test
Train= df.drop('variety',axis=1)
Test=df1['variety']


#input:90% để train và 10% để test
tp=int(0.9*len(Train))
X_train=Train[:tp]
Y_train=Test[:tp]
X_test=Train[tp:]
Y_test=Test[tp:]
#xây dựng mô hình
from sklearn.linear_model import RidgeClassifier
reg=RidgeClassifier().fit(X_train,Y_train)
pred=reg.predict(X_test)
reg.score(X_test,Y_test)
target_class=pd.DataFrame( {'Test Value':pred , 'True Value' : Y_test} )
target_class
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y_test, pred)
print("Accuracy:", accuracy)

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:, :-1], df.iloc[:, -1], test_size=0.1, random_state=42)
```

# **Thử mấy model dùng để classify để xem model nào classify ra accuracy cao nhất**

1.**KNN**

```python
from sklearn import neighbors
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python

clf = neighbors.KNeighborsClassifier(n_neighbors = 10, weights = 'distance', metric='euclidean')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print("KNN accuracy: ", accuracy_score(y_test, y_pred))
print("KNN f1 score: ", f1_score(y_test, y_pred, average='weighted'))
print("KNN confusion matrix: ")
print(confusion_matrix(y_test, y_pred))
```

2.**Logistic regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
regressor = LogisticRegression()
regressor.fit(X_train, y_train)

# predict
predictions = regressor.predict(X_test)
# evaluate
print("LR classification accuracy", accuracy_score(y_test, predictions))

# matrix confusion
cm = confusion_matrix(y_test, predictions)
print(cm)

# f1 score
print("LR f1 score", f1_score(y_test, predictions, average="macro"))
```
